```python
from zhipuai import ZhipuAI
import os
# 使用前配置自己的 api 到环境变量中如
import sys
from dotenv import load_dotenv, find_dotenv
# _ = load_dotenv(find_dotenv())
_ = load_dotenv('../QA_Project/.env')

# 获取环境变量 
zhipuai_api_key = os.environ["ZHIPUAI_API_KEY"]
```

```python
client = ZhipuAI(api_key=zhipuai_api_key) 
response = client.embeddings.create(
    model="embedding-2", #填写需要调用的模型名称
    input="你好",
)
```

```python
response
```

```python
print(f"生成的 embedding 长度为: {len(response.data[0].embedding)}")
```

```python
print(f"用户输入的 tokens 数量为: {response.usage.prompt_tokens}")
```

```python
print(f"总 tokens 数量为: {response.usage.total_tokens}")
```

```python
from QA_Project.project.embedding.zhipuai_embedding import ZhipuAIEmbeddings
```

```python
zhipuai_embeddings = ZhipuAIEmbeddings(zhipuai_api_key=zhipuai_api_key)
```

```python
query_embedding = zhipuai_embeddings.embed_query('你好')
print(query_embedding[:10])
```

```python
doc_list = [
    '你好',
    '什么是机器学习',
    '什么是深度学习',
    '什么是大模型'
]
doc_embeddings = zhipuai_embeddings.embed_documents(doc_list)
print('\n'.join([f'{doc} 的 embedding 为 {doc_embedding[:10]}...\n' for doc, 
                 doc_embedding in zip(doc_list, doc_embeddings)]))
```
